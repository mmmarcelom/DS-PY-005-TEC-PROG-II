```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tips = sns.load_dataset('tips')
```

#### Questão 1

Qual o valor médio da gorjeta (*tip*) de acordo com o dia da semana (*day*)?

```python
sns.barplot(data=df_tips, x='day', y='tip', estimator='mean', errorbar=None)
plt.show()
```

#### Questão 2

Qual a soma das gorjetas (*tip*) por turno do dia (*time*), observando os fumantes e não fumantes (*smoker*)?

```python
sns.barplot(data=df_tips, x='day', y='tip', hue='smoker', estimator='sum', errorbar=None)
plt.show()
```
